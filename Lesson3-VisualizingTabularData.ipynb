 ### Visualizing Data through Matplotlib

```python
#importing library

import numpy
import matplotlib.pyplot

#importing data

data = numpy.loadtxt(fname='python-novice-inflammation-data/data/inflammation-01.csv', delimiter=',')
```

```python
image = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.show()
```

Each row in the `heat map` represents to a patient in the dataset, and each column corresponds to a day in the dataset. `Blue pixel` in the heat map represent low values, and `yellow pixel` represents high values.  
  
From this heat map, we can conclude that  
1. Patients took their medication once their inflammation flare-ups begin.
2. It takes around 3 weeks for medication to take effect and begin reducing flare-ups.
3. And flare-ups appear to drop to zero by the end of the clinical trial.


```python
ave_inflammation = numpy.mean(data, axis=0)
ave_plot = matplotlib.pyplot.plot(ave_inflammation)
matplotlib.pyplot.show()
```

This `line graph` show the average inflammation per day across all patient. The graph support the claim that the medication need 3 weeks to work.

```python
max_plot = matplotlib.pyplot.plot(numpy.max(data, axis=0))
matplotlib.pyplot.show()
```

```python
min_plot = matplotlib.pyplot.plot(numpy.min(data, axis=0))
matplotlib.pyplot.show()
```

The maximum value rises and falls linearly, while the minimum seems to be a step function. **Neither trend seems particularly likely**, so either there's a mistake in our calculations or something wrong with our data.

### Grouping plots

Group similar plots in a single figure using subplots.

```python
fig = matplotlib.pyplot.figure(figsize = (10.0, 3.0)) 
# creates a space for plots
# "figsize" tells python how big to make this space

axes1 = fig.add_subplot(1, 3, 1) # total rows, total number of column, referencing
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

# subplot stored in variable axes1, axes2 and axes3
# add subplot parameter place subplot to figure, first parameter denotes how many total rows of subplot,
# second parameter refers to toal number of subplot columns, third parameter denotes which subplot your
# variable is referencing (left to right or top to bottom)

axes1.set_ylabel('average') #set the name of y label
axes1.plot(numpy.mean(data, axis=0))

axes2.set_ylabel('max')
axes2.plot(numpy.mean(data, axis=0))

axes3.set_ylabel('min')
axes3.plot(numpy.min(data, axis=0))

axes3.set_ylim(0,6) # modify the limit of y line.

# other options: name the x label with  set_xlabel()

fig.tight_layout()

matplotlib.pyplot.savefig('img/inflammation.png') # save the inflammation subplots to the folder img/
matplotlib.pyplot.show()
```

`matplotlib` normally sets x and y axes limits to the min and max of the data.

```python

fig2 = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

axes4 = fig2.add_subplot(1,3,1)
axes5 = fig2.add_subplot(1,3,2)
axes6 = fig2.add_subplot(1,3,3)

axes4.set_ylabel('average')
axes4.plot(numpy.mean(data, axis=0), drawstyle='steps-mid')

axes5.set_ylabel('max')
axes5.plot(numpy.mean(data, axis=0), drawstyle='steps-mid')

axes6.set_ylabel('min')
axes6.plot(numpy.min(data, axis=0), drawstyle='steps-mid')

fig2.tight_layout()
matplotlib.pyplot.show()
```
